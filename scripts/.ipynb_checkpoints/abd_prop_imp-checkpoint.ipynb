```python
#partition 49 patients to 39 train 10 test - random seed done once
trainIDs=[25, 34, 6, 12, 21, 39, 8, 29, 13, 45, 36, 19, 20, 3, 26, 18, 24, 27, 1, 46, 11, 5, 10, 23, 30, 0, 47, 41, 14, 22, 31, 35, 37, 48, 15, 17, 4, 32, 44]
testIDs=[]
for i in range(49):
    if i not in trainIDs:
        testIDs.append(i)
print(trainIDs)
print(testIDs)
```

```python
max_len=0
min_len=1400
for i in trainIDs:
    subjectID = PARTICIPANT_IDS[i]
    sleep = read_patient_data(subjectID)
    length=len(sleep["sleep_stage_num_psg"].values)
    if max_len<length:
        max_len=length
    if min_len>length:
        min_len=length
print(max_len,min_len)
```

```python
emfit_vals=np.zeros((1491,39))
somnofy_vals=np.zeros((1491,39))
psg_vals=np.zeros((1491,39))
for i in range(len(trainIDs)):
    s_stages = read_patient_data(PARTICIPANT_IDS[trainIDs[i]])
    rows=len(s_stages["sleep_stage_num_psg"].values)
    n_loop=max_len-rows
    if n_loop!=0:
        per_data=math.floor(rows/n_loop)
        if per_data==0:
            per_data=1
    i_th=0
    increment="5"
    for j in range(n_loop):
        if i_th >= rows:
            i_th=0
            increment='8'
        a=s_stages.iloc[i_th]
        row=s_stages.index.values[i_th]
        row=row[:18]+increment+row[19:]
        s_stages.loc[row]=a
        i_th+=per_data
    sorted_sleep=s_stages.sort_index()
    emfit_vals[:,i]=sorted_sleep["sleep_stage_num_emfit"].values
    somnofy_vals[:,i]=sorted_sleep["sleep_stage_num_somnofy"].values
    psg_vals[:,i]=sorted_sleep["sleep_stage_num_psg"].values   
```

# Approach 2

```python
loss=100
num_of_update=0
w_ap2=np.zeros((2,1), dtype=float)
w = np.zeros((2,1), dtype=float)
step=0.002
while w[0] <=0.6:
    w[1]=0
    while w[1] <=0.6:
        e=np.zeros((1491,39))
        for i in range(39):
            e[:,i]=(psg_vals[:,i].reshape(1491,1)-np.c_[emfit_vals[:,i],somnofy_vals[:,i]].reshape(1491,2)@w).reshape(1491,)
        mse=np.zeros((39,1))
        for i in range(39):
            mse[i]=1/2/1491*e[:,i].T@e[:,i]
        average_mse=np.mean(mse)
        if average_mse<loss:
            num_of_update+=1
            loss=average_mse
            w_ap2[0]=w[0]
            w_ap2[1]=w[1]
        w[1]+=step
    w[0]+=step
```

```python
def ap3_build_ss(emfit,somnofy,avrg_acc_somnofy,std_somnofy,avrg_acc_emfit,std_emfit):
    ss_a3=np.zeros(len(emfit))

    for j in range(100):
        for i in range(len(emfit)):
            gaus_acc_somnofy=random.gauss(avrg_acc_somnofy[int(somnofy[i])],std_somnofy[int(somnofy[i])])
            gaus_acc_emfit=random.gauss(avrg_acc_emfit[int(emfit[i])],std_emfit[int(emfit[i])])
            value=gaus_acc_somnofy/(gaus_acc_somnofy+gaus_acc_emfit)
            rand_num=random.randint(0,100)

            if rand_num/100 <= value:
                ss_a3[i]+=somnofy[i]
            else:
                ss_a3[i]+=emfit[i]
    for i in range(len(psg)):
        ss_a3[i]=round(ss_a3[i]/100)
    return(ss_a3)   
```

```python
def build_new_ss_methodx(weights,ss_emfit,ss_somnofy):
    new_SS=np.ones(len(ss_emfit))
    for i in range(len(ss_emfit)):
        if ss_emfit[i]==1:
            if ss_somnofy[i]==1:
                new_SS[i]=ss_emfit[i]-1
            elif ss_somnofy[i]==2:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[0][0]+ss_somnofy[i]*weights[0][1]),True)
            elif ss_somnofy[i]==3:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[1][0]+ss_somnofy[i]*weights[1][1]),True)
            elif ss_somnofy[i]==4:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[2][0]+ss_somnofy[i]*weights[2][1]),True)
        elif ss_emfit[i]==2:
            if ss_somnofy[i]==1:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[3][0]+ss_somnofy[i]*weights[3][1]),True)
            elif ss_somnofy[i]==2:
                new_SS[i]=ss_emfit[i]-1
            elif ss_somnofy[i]==3:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[4][0]+ss_somnofy[i]*weights[4][1]),True)
            elif ss_somnofy[i]==4:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[5][0]+ss_somnofy[i]*weights[5][1]),True)
        elif ss_emfit[i]==3:
            if ss_somnofy[i]==1:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[6][0]+ss_somnofy[i]*weights[6][1]),True)
            elif ss_somnofy[i]==2:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[7][0]+ss_somnofy[i]*weights[7][1]),True)
            elif ss_somnofy[i]==3:
                new_SS[i]=ss_emfit[i]-1
            elif ss_somnofy[i]==4:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[8][0]+ss_somnofy[i]*weights[8][1]),True)
        elif ss_emfit[i]==4:
            if ss_somnofy[i]==1:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[9][0]+ss_somnofy[i]*weights[9][1]),True)
            elif ss_somnofy[i]==2:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[10][0]+ss_somnofy[i]*weights[10][1]),True)
            elif ss_somnofy[i]==3:
                new_SS[i]=value_rounded((ss_emfit[i]*weights[11][0]+ss_somnofy[i]*weights[10][1]),True)
            elif ss_somnofy[i]==4:
                new_SS[i]=ss_emfit[i]-1
    return new_SS
```
