```python
acc_per_somnofy_ss0=np.zeros(49)
acc_per_somnofy_ss1=np.zeros(49)
acc_per_somnofy_ss2=np.zeros(49)
acc_per_somnofy_ss3=np.zeros(49)
acc_per_emfit_ss0=np.zeros(49)
acc_per_emfit_ss1=np.zeros(49)
acc_per_emfit_ss2=np.zeros(49)
acc_per_emfit_ss3=np.zeros(49)

for i in range(len(PARTICIPANT_IDS)):
    subjectID = PARTICIPANT_IDS[i]
    ss_patient = read_patient_data(subjectID)
    y_true=ss_patient["sleep_stage_num_psg"]
    
    for k in range(0,4):
        ss=ss_patient.loc[(ss_patient["sleep_stage_num_somnofy"]==k)]
        if k == 0:
            if not ss.empty:
                acc_per_somnofy_ss0[i]=accuracy_score(ss["sleep_stage_num_psg"],ss["sleep_stage_num_somnofy"])
            else:
                if k not in y_true.values:                    
                    acc_per_somnofy_ss0[i]=1
                else:                     
                    acc_per_somnofy_ss0[i]=0
        elif k == 1:
            if not ss.empty:
                acc_per_somnofy_ss1[i]=accuracy_score(ss["sleep_stage_num_psg"],ss["sleep_stage_num_somnofy"])
            else:
                if k not in y_true.values:                    
                    acc_per_somnofy_ss1[i]=1
                else:                     
                    acc_per_somnofy_ss1[i]=0
        elif k == 2:
            if not ss.empty:
                acc_per_somnofy_ss2[i]=accuracy_score(ss["sleep_stage_num_psg"],ss["sleep_stage_num_somnofy"])
            else:
                if k not in y_true.values:                    
                    acc_per_somnofy_ss2[i]=1
                else:                     
                    acc_per_somnofy_ss2[i]=0
        elif k == 3:
            if not ss.empty:
                acc_per_somnofy_ss3[i]=accuracy_score(ss["sleep_stage_num_psg"],ss["sleep_stage_num_somnofy"])
            else:
                if k not in y_true.values:                    
                    acc_per_somnofy_ss3[i]=1
                else:                     
                    acc_per_somnofy_ss3[i]=0
    
    for j in range(0,4):
        ss=ss_patient.loc[(ss_patient["sleep_stage_num_emfit"]==j)]
        if j == 0:
            if not ss.empty:
                acc_per_emfit_ss0[i]=accuracy_score(ss["sleep_stage_num_psg"],ss["sleep_stage_num_emfit"])
            else:
                if j not in y_true.values:                    
                    acc_per_emfit_ss0[i]=1
                else:                     
                    acc_per_emfit_ss0[i]=0
        elif j == 1:
            if not ss.empty:
                acc_per_emfit_ss1[i]=accuracy_score(ss["sleep_stage_num_psg"],ss["sleep_stage_num_emfit"])
            else:
                if j not in y_true.values:                    
                    acc_per_emfit_ss1[i]=1
                else:                     
                    acc_per_emfit_ss1[i]=0  
        elif j == 2:
            if not ss.empty:
                acc_per_emfit_ss2[i]=accuracy_score(ss["sleep_stage_num_psg"],ss["sleep_stage_num_emfit"])
            else:
                if j not in y_true.values:                    
                    acc_per_emfit_ss2[i]=1
                else:                     
                    acc_per_emfit_ss2[i]=0
        elif j == 3:
            if not ss.empty:
                acc_per_emfit_ss3[i]=accuracy_score(ss["sleep_stage_num_psg"],ss["sleep_stage_num_emfit"])
            else:
                if j not in y_true.values:                    
                    acc_per_emfit_ss3[i]=1
                else:                     
                    acc_per_emfit_ss3[i]=0
```

```python
print("Somnofy average accuracy for 0 :",np.average(acc_per_somnofy_ss0), "Std: ", np.std(acc_per_somnofy_ss0))
print("Somnofy average accuracy for 1 :",np.average(acc_per_somnofy_ss1), "Std: ", np.std(acc_per_somnofy_ss1))
print("Somnofy average accuracy for 2 :",np.average(acc_per_somnofy_ss2), "Std: ", np.std(acc_per_somnofy_ss2))
print("Somnofy average accuracy for 3 :",np.average(acc_per_somnofy_ss3), "Std: ", np.std(acc_per_somnofy_ss3))
print("Emfit average accuracy for 0 :",np.average(acc_per_emfit_ss0), "Std: ", np.std(acc_per_emfit_ss0))
print("Emfit average accuracy for 1 :",np.average(acc_per_emfit_ss1), "Std: ", np.std(acc_per_emfit_ss1))
print("Emfit average accuracy for 2 :",np.average(acc_per_emfit_ss2), "Std: ", np.std(acc_per_emfit_ss2))
print("Emfit average accuracy for 3 :",np.average(acc_per_emfit_ss3), "Std: ", np.std(acc_per_emfit_ss3))
```

```python
avrg_acc_somnofy=[np.average(acc_per_somnofy_ss0),np.average(acc_per_somnofy_ss1),np.average(acc_per_somnofy_ss2),np.average(acc_per_somnofy_ss3)]
std_somnofy=[np.std(acc_per_somnofy_ss0),np.std(acc_per_somnofy_ss1),np.std(acc_per_somnofy_ss2),np.std(acc_per_somnofy_ss3)]
avrg_acc_emfit=[np.average(acc_per_emfit_ss0),np.average(acc_per_emfit_ss1),np.average(acc_per_emfit_ss2),np.average(acc_per_emfit_ss3)]
std_emfit=[np.std(acc_per_emfit_ss0),np.std(acc_per_emfit_ss1),np.std(acc_per_emfit_ss2),np.std(acc_per_emfit_ss3)]
```

```python
ss_m6=np.zeros(len(sleep_stages["sleep_stage_num_emfit"]))

for i in range(len(sleep_stages["sleep_stage_num_emfit"])):
    gaus_acc_somnofy=random.gauss(avrg_acc_somnofy[int(sleep_stages["sleep_stage_num_somnofy"][i])],std_somnofy[int(sleep_stages["sleep_stage_num_somnofy"][i])])
    gaus_acc_emfit=random.gauss(avrg_acc_emfit[int(sleep_stages["sleep_stage_num_emfit"][i])],std_emfit[int(sleep_stages["sleep_stage_num_emfit"][i])])
    value=gaus_acc_somnofy/(gaus_acc_somnofy+gaus_acc_emfit)
    rand_num=random.randint(0,100)
    
    if rand_num/100 <= value:
        ss_m6[i]=sleep_stages["sleep_stage_num_somnofy"][i]
    else:
        ss_m6[i]=sleep_stages["sleep_stage_num_emfit"][i]
```

```python
conf_somnofy=np.zeros((4,4))
conf_emfit=np.ones((4,4))
for i in range(len(PARTICIPANT_IDS)):
    subjectID = PARTICIPANT_IDS[i]
    ss_patients = read_patient_data(subjectID)
    psg=ss_patients["sleep_stage_num_psg"]
    
    conf_somnofy+=confusion_matrix(psg,ss_patients["sleep_stage_num_somnofy"])
    conf_emfit+=confusion_matrix(psg,ss_patients["sleep_stage_num_emfit"])
print(conf_somnofy/49)
print(conf_emfit/49)

for i in range(len(conf_somnofy)):
    for j in range(len(conf_somnofy[0])):
        conf_somnofy[i][j]=round(conf_somnofy[i][j]/49)
for i in range(len(conf_emfit)):
    for j in range(len(conf_emfit[0])):
        conf_emfit[i][j]=round(conf_emfit[i][j]/49)
print(conf_somnofy)
print(conf_emfit)
```

```python
when_emfit_0=[]
when_emfit_1=[]
when_emfit_2=[]
when_emfit_3=[]
when_somnofy_0=[]
when_somnofy_1=[]
when_somnofy_2=[]
when_somnofy_3=[]

for i in range(4):
    sum=0
    for j in range(4):
        sum+=conf_emfit[j][i]
    for j in range(4):
        if i==0:
            when_emfit_0.append(conf_emfit[j][i]/sum)
        elif i==1:
            when_emfit_1.append(conf_emfit[j][i]/sum)
        elif i==2:
            when_emfit_2.append(conf_emfit[j][i]/sum)
        elif i==3:
            when_emfit_3.append(conf_emfit[j][i]/sum)
        
for i in range(4):
    sum=0
    for j in range(4):
        sum+=conf_somnofy[j][i]
    for j in range(4):
        if i==0:
            when_somnofy_0.append(conf_somnofy[j][i]/sum)
        elif i==1:
            when_somnofy_1.append(conf_somnofy[j][i]/sum)
        elif i==2:
            when_somnofy_2.append(conf_somnofy[j][i]/sum)
        elif i==3:
            when_somnofy_3.append(conf_somnofy[j][i]/sum)
```

```python
balanced_acc_m1=[]
balanced_acc_m2=[]
balanced_acc_m3=[]
balanced_acc_m4=[]
balanced_acc_m5=[]
balanced_acc_m6=[]
balanced_acc_radar=[]
balanced_acc_mat=[]

for i in range(len(PARTICIPANT_IDS)):
    subjectID = PARTICIPANT_IDS[i]
    stages = read_patient_data(subjectID)
    psg_val = stages["sleep_stage_num_psg"]
    
    w_m1_l1,loss_m1_l1=mse_gd(stages["sleep_stage_num_emfit"].to_numpy(),stages["sleep_stage_num_somnofy"].to_numpy(),stages["sleep_stage_num_psg"].to_numpy(), max_iters=1000, gamma=0.01)
    ss_m1_l1=build_new_ss(w_m1,stages["sleep_stage_num_emfit"],stages["sleep_stage_num_somnofy"])
    
    w_m2_l1,loss_m2_l1=mse_gd(stages["sleep_stage_num_emfit"].to_numpy(),stages["sleep_stage_num_somnofy"].to_numpy(),stages["sleep_stage_num_psg"].to_numpy(), max_iters=1500, gamma=0.01,w0=True)
    ss_m2_l1=build_new_ss(w_m2,stages["sleep_stage_num_emfit"],stages["sleep_stage_num_somnofy"])   
    
    w_m3_l1,loss_m3_l1=mse_gd((stages["sleep_stage_num_emfit"]+1).to_numpy(),(stages["sleep_stage_num_somnofy"]+1).to_numpy(),(stages["sleep_stage_num_psg"]+1).to_numpy(), max_iters=2000, gamma=0.01,w0=False)
    ss_m3_l1=build_new_ss(w_m3,(stages["sleep_stage_num_emfit"]+1),(stages["sleep_stage_num_somnofy"]+1),False,True)
    
    w_m4_l1,loss_m4_l1=mse_gd((stages["sleep_stage_num_emfit"]+1).to_numpy(),(stages["sleep_stage_num_somnofy"]+1).to_numpy(),(stages["sleep_stage_num_psg"]+1).to_numpy(), max_iters=2000, gamma=0.01,w0=True)
    ss_m4_l1=build_new_ss(w_m4,(stages["sleep_stage_num_emfit"]+1),(stages["sleep_stage_num_somnofy"]+1),True,True)
    
    w_m5_l1=method_x(stages+1, max_iters=150, gamma=0.005)
    ss_m5_l1=build_new_ss_methodx(w_m5,(stages["sleep_stage_num_emfit"]+1),(stages["sleep_stage_num_somnofy"]+1))
    
    ss_m6_l1=np.zeros(len(stages["sleep_stage_num_emfit"]))

    for i in range(len(stages["sleep_stage_num_emfit"])):
        gaus_acc_somnofy=random.gauss(avrg_acc_somnofy[int(stages["sleep_stage_num_somnofy"][i])],std_somnofy[int(stages["sleep_stage_num_somnofy"][i])])
        gaus_acc_emfit=random.gauss(avrg_acc_emfit[int(stages["sleep_stage_num_emfit"][i])],std_emfit[int(stages["sleep_stage_num_emfit"][i])])
        value=gaus_acc_somnofy/(gaus_acc_somnofy+gaus_acc_emfit)
        rand_num=random.randint(0,100)
        if rand_num/100 <= value:
            ss_m6_l1[i]=stages["sleep_stage_num_somnofy"][i]
        else:
            ss_m6_l1[i]=stages["sleep_stage_num_emfit"][i]
    
    balanced_acc_m1.append(balanced_accuracy_score(psg_val,ss_m1_l1))
    balanced_acc_m2.append(balanced_accuracy_score(psg_val,ss_m2_l1))
    balanced_acc_m3.append(balanced_accuracy_score(psg_val,ss_m3_l1))
    balanced_acc_m4.append(balanced_accuracy_score(psg_val,ss_m4_l1))
    balanced_acc_m5.append(balanced_accuracy_score(psg_val,ss_m5_l1))
    balanced_acc_m6.append(balanced_accuracy_score(psg_val,ss_m6_l1))
    balanced_acc_radar.append(balanced_accuracy_score(psg_val,stages["sleep_stage_num_somnofy"]))
    balanced_acc_mat.append(balanced_accuracy_score(psg_val,stages["sleep_stage_num_emfit"]))
    
balanced_acc_m1=np.asarray(balanced_acc_m1)
balanced_acc_m2=np.asarray(balanced_acc_m2)
balanced_acc_m3=np.asarray(balanced_acc_m3)
balanced_acc_m4=np.asarray(balanced_acc_m4)
balanced_acc_m5=np.asarray(balanced_acc_m5)
balanced_acc_m6=np.asarray(balanced_acc_m6)
balanced_balanced_acc_radar=np.asarray(balanced_acc_radar)
balanced_acc_mat=np.asarray(balanced_acc_mat)
```
